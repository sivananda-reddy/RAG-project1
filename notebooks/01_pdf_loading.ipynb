## Load Sample Documents

Let's start with sample documents to understand the structure before working with real PDFs.

```python
# Load sample documents
documents = load_sample_documents()

print(f"Loaded {len(documents)} sample documents\n")

# Examine the structure
for i, doc in enumerate(documents):
    print(f"Document {i+1}:")
    print(f"  Filename: {doc['filename']}")
    print(f"  Pages: {doc['num_pages']}")
    print(f"  Text length: {len(doc['text'])} characters")
    print(f"  Preview: {doc['text'][:100]}...\n")
```

## Load Real PDFs from data/ Directory

```python
# Initialize PDF loader
loader = PDFLoader(data_path="../data")

# Try to load PDFs from data directory
pdf_documents = loader.load_all_pdfs()

if pdf_documents:
    print(f"✓ Loaded {len(pdf_documents)} PDFs from data directory\n")
    for doc in pdf_documents:
        print(f"  - {doc['filename']}: {len(doc['text'])} characters, {doc.get('num_pages', '?')} pages")
else:
    print("ℹ No PDFs found in data/ directory")
    print("  Tip: Copy your PDF files to the data/ folder and run this cell again")
```

## Text Chunking

Large documents need to be split into smaller, manageable chunks for embeddings.

```python
# Use sample document for chunking demo
sample_text = documents[0]['text']

# Split text into chunks
chunks = loader.split_text(sample_text, chunk_size=200, overlap=50)

print(f"Original text: {len(sample_text)} characters")
print(f"Number of chunks: {len(chunks)}")
print(f"Average chunk size: {len(sample_text) // len(chunks)} characters\n")

print("First 3 chunks:")
for i, chunk in enumerate(chunks[:3]):
    print(f"\nChunk {i+1} ({len(chunk)} chars):")
    print(chunk)
```

## Understanding Chunk Size and Overlap

Let's visualize how different chunk sizes affect the chunking process.

```python
# Test different chunk sizes
test_text = "This is a test document. " * 20  # Create sample text

chunk_sizes = [100, 200, 500]

for chunk_size in chunk_sizes:
    chunks = loader.split_text(test_text, chunk_size=chunk_size, overlap=50)
    print(f"Chunk size: {chunk_size}")
    print(f"  Number of chunks: {len(chunks)}")
    print(f"  First chunk: {chunks[0][:50]}...")
    print()
```

## Document Statistics

```python
import statistics

# Analyze document statistics
all_docs = documents + pdf_documents if pdf_documents else documents

doc_lengths = [len(doc['text']) for doc in all_docs]

print("Document Statistics:")
print(f"  Total documents: {len(all_docs)}")
print(f"  Average length: {statistics.mean(doc_lengths):.0f} characters")
print(f"  Min length: {min(doc_lengths)} characters")
print(f"  Max length: {max(doc_lengths)} characters")
print(f"  Total content: {sum(doc_lengths)} characters")
```
